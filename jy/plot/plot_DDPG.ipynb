```python
import numpy as np
import matplotlib.pyplot as plt

false_random_data=np.load("../results/DDPG_False_0_moving_training_random.npy")
f_sum_random=false_random_data[-1]
false_load_data=np.load("../results/DDPG_False_0_moving_training.npy")+f_sum_random
false_load_data=np.concatenate((false_random_data,false_load_data))

true_random_data=np.load("../results/DDPG_True_0_moving_training_random.npy")
t_sum_random=true_random_data[-1]
true_load_data=np.load("../results/DDPG_True_0_moving_training.npy")+t_sum_random
true_load_data=np.concatenate((true_random_data,true_load_data))

fig=plt.figure(figsize=(25,8))
plt.plot(true_load_data,label="with OIAC")
plt.plot(false_load_data,label="without OIAC")
plt.legend()
plt.show()
```
