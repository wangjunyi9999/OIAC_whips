```python
import numpy as np
import matplotlib.pyplot as plt

mat_k=np.load('../tmp_k_True.npy')
mat_k_false=np.load('../tmp_k_False.npy')
mat_b=np.load('../tmp_b_True.npy')
mat_b_false=np.load('../tmp_b_False.npy')
mat_v=np.load('../tmp_v_True.npy')
mat_v_false=np.load('../tmp_v_False.npy')
mat_p=np.load('../tmp_p_True.npy')
mat_p_false=np.load('../tmp_p_False.npy')

shape=mat_v.shape[0]
mat_v=mat_v.reshape(shape,2)
mat_p=mat_p.reshape(shape,2)

shape_f=mat_v_false.shape[0]
mat_v_false=mat_v_false.reshape(shape_f,2)
mat_p_false=mat_p_false.reshape(shape_f,2)

k00=mat_k[:,0][:,0]
k11=mat_k[:,1][:,1]
k00_f=mat_k_false[:,0][:,1]
k11_f=mat_k_false[:,1][:,0]


b00=mat_b[:,0][:,0]
b11=mat_b[:,1][:,1]
b00_f=mat_b_false[:,0][:,1]
b11_f=mat_b_false[:,1][:,0]

v1=mat_v[:,0]
v2=mat_v[:,1]
p1=mat_p[:,0]
p2=mat_p[:,1]

v1_f=mat_v_false[:,0]
v2_f=mat_v_false[:,1]
p1_f=mat_p_false[:,0]
p2_f=mat_p_false[:,1]


width=15
height=8
fig,axes= plt.subplots(2,2, figsize=(width,height))  
axes[0,0].plot(k00)
axes[0,0].plot(b00)
axes[0,0].plot(p1)
axes[0,0].plot(v1)
axes[0,0].set_title("joint_0_oiac")


axes[0,1].plot(k00_f)
axes[0,1].plot(b00_f)
axes[0,1].plot(p1_f)
axes[0,1].plot(v1_f)
axes[0,1].set_title("joint_0")

axes[1,1].plot(k11_f)
axes[1,1].plot(b11_f)
axes[1,1].plot(p2_f)
axes[1,1].plot(v2_f)
axes[1,1].set_title("joint_1")

axes[1,0].plot(k11,label="k")
axes[1,0].plot(b11,label="b")
axes[1,0].plot(p2, label="q1_err")
axes[1,0].plot(v2, label="dq1_err")
axes[1,0].set_title("joint_1_oiac")
fig.legend()
plt.show()
```

```python
# comparsion bar chart
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font',family='Times New Roman')
# mpl.rcParams['font.weight'] = 'bold'
#mpl.rcParams['font.sans-serif'] = ['Times New Roman']  # tmd no Roman in the system !!!
width=28
height=15
bwith = 6 #6
line_width=9
font_size=40
alpha=0.6 # for the transparent
def add_labels(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2, height, height, ha='center', va='bottom', size=font_size)
        rect.set_edgecolor('white')

X = ['pos_err_s','pos_err_e','vel_err_s','vel_err_e']
Cvalue = [0.161,0.038,1.038,0.229]
Ovalue = [0.036,0.009,0.138,0.073]
w_width=0.4# if n sub bar plot, w_width should be smaller than 1/n
X_axis = np.arange(len(X))

fig,axe=plt.subplots(1,1,figsize=(width,height))
# fig,axe=plt.subplots()
rects1 = axe.bar(X_axis - w_width/2, Cvalue, w_width, label = 'CONSTANT', facecolor='deeppink',alpha=alpha)
rects2 = axe.bar(X_axis + w_width/2, Ovalue, w_width, label='OIAC', facecolor = 'darkblue',alpha=alpha)

ax=plt.gca()
ax.spines['top'].set_color('none')  # set the upper spine as none
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_linewidth(bwith)
ax.spines['left'].set_linewidth(bwith)
plt.xticks(X_axis, X)


axe.tick_params(axis='x',labelsize=font_size)
axe.tick_params(axis='y',labelsize=font_size)

# axe.bar_label(rects1, padding=3)
# axe.bar_label(rects2, padding=3)

add_labels(rects1)
add_labels(rects2)
# plt.xlabel("Groups")
plt.ylabel("[rad]", loc='top', fontsize=font_size, rotation=0)
fig.tight_layout()
plt.legend(loc='upper right',prop={'size':font_size})
#plt.savefig("../images/compare_err.png",dpi=300)
plt.show()
```
