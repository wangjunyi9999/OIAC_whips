```python
import numpy as np
import matplotlib.pyplot as plt
# hit 390 times
true_load_data=np.load("../results/DDPG_True_training.npy") # blue one 
random_load_data=np.load("../results/DDPG_True_training_random.npy")
true_load_data=np.concatenate((random_load_data,true_load_data))

false_load_data=np.load("../results/DDPG_False_training.npy")
false_random_data=np.load("../results/DDPG_False_training_random.npy")# orange one
false_load_data=np.concatenate((false_random_data,false_load_data))

fig=plt.figure(figsize=(25,8))
plt.plot(true_load_data,label="with OIAC")
plt.plot(false_load_data,label="without OIAC")
plt.legend()
plt.show()
```
