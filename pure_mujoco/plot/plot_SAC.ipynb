```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
plt.rc('font',family='DeJavu Serif')
mpl.rcParams['font.weight'] = 'bold'
width=38
height=15
bwith = 6 #2
line_width=9
font_size=40
false_random_data_1=np.load("../results/SAC_False_40_moving_training_random.npy")
f_sum_random_1=false_random_data_1[-1]
false_load_data_1=np.load("../results/SAC_False_40_moving_training.npy")+f_sum_random_1
false_load_data_1=np.concatenate((false_random_data_1,false_load_data_1))

true_random_data_1=np.load("../results/SAC_True_40_moving_training_random.npy")
t_sum_random_1=true_random_data_1[-1]
true_load_data_1=np.load("../results/SAC_True_40_moving_training.npy")+t_sum_random_1
true_load_data_1=np.concatenate((true_random_data_1,true_load_data_1))

false_random_data=np.load("../results/SAC_False_30_moving_training_random.npy")
f_sum_random=false_random_data[-1]
false_load_data=np.load("../results/SAC_False_30_moving_training.npy")+f_sum_random
false_load_data=np.concatenate((false_random_data,false_load_data))

true_random_data=np.load("../results/SAC_True_30_moving_training_random.npy")
t_sum_random=true_random_data[-1]
true_load_data=np.load("../results/SAC_True_30_moving_training.npy")+t_sum_random
true_load_data=np.concatenate((true_random_data,true_load_data))

# true_size=min(true_load_data_1.shape[0],true_load_data.shape[0])
# false_size=min(false_load_data_1.shape[0], false_load_data.shape[0])

true_size=121
false_size=121

sns_false_data=np.vstack((false_load_data[:false_size],false_load_data_1[:false_size]))
sns_true_data=np.vstack((true_load_data[:true_size],true_load_data_1[:true_size]))
df_false= pd.DataFrame(sns_false_data).melt(var_name='episode',value_name='average_reward')
df_true=pd.DataFrame(sns_true_data).melt(var_name='episode',value_name='average_reward')
# df=[df_true,df_false]


# plt.figure(figsize=(width,height))
fig,axes=plt.subplots(1,1,figsize=(width,height))
axes.tick_params(axis='x',labelsize=font_size)
axes.tick_params(axis='y',labelsize=font_size)
axes.set_ylabel('Reward',fontsize=font_size,loc='top', rotation=0)
axes.set_xlabel('Episode [1e2]',fontsize=font_size,loc='right', rotation=0)

sns.lineplot(x="episode", y="average_reward", label='CONSTANT', data=df_false,linewidth=line_width)
sns.lineplot(x="episode", y="average_reward", label='OIAC',data=df_true,linewidth=line_width)

plt.legend(loc='upper right',prop={'size':font_size})
plt.savefig("../images/SAC.png",dpi=300)
plt.show()

```
