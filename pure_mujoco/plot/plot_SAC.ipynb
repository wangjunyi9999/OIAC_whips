```python
t_set=[]
tt_set=[]
set_value=100.0
for i in range(len(true_load_data)):
    if i % 100==99:
        t_set.append(true_load_data[i]-true_load_data[0])


for j in range(len(t_set)-1):        
        tt_set.append((t_set[j+1]-t_set[j])/set_value)
tt_set.insert(0,t_set[0]/set_value)


f_set=[]
ff_set=[]
set_value=100.0
for i in range(len(false_load_data)):
    if i % 100==99:
        f_set.append(false_load_data[i]-false_load_data[0])


for j in range(len(f_set)-1):        
        ff_set.append((f_set[j+1]-f_set[j])/set_value)
ff_set.insert(0,f_set[0]/set_value)

fig=plt.figure(figsize=(12,8))
plt.plot(tt_set)
plt.plot(ff_set)

```
