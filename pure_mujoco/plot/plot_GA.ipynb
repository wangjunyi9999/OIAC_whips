```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.rc('font',family='DeJavu Serif')
mpl.rcParams['font.weight'] = 'bold'
width=38
height=15
bwith = 6 #边框宽度设置为2
line_width=9
font_size=40
false_load_data=np.load("../GA_res/GA_False_result.npy")
true_load_data=np.load("../GA_res/GA_True_result.npy")
fig,axes=plt.subplots(1,1,figsize=(width,height))
ax = plt.gca()

# 隐藏右边和顶部的线条
ax.spines['top'].set_color('none')  # set the upper spine as none
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_linewidth(bwith)
ax.spines['left'].set_linewidth(bwith)

# 设置x轴和y轴所在的位置
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# 设置x轴和y轴的交点为(0,0)
ax.spines['bottom'].set_position(('data', 0.03))
ax.spines['left'].set_position(('data', 0))
axes.set_ylabel('D [m]',fontsize=font_size,loc='top', rotation=0)
axes.set_xlabel('Episode Time',fontsize=font_size,loc='right', rotation=0)
axes.plot(false_load_data,label="CONSTANT",linewidth=line_width, color='#32B897')
axes.plot(true_load_data,label="OIAC",linewidth=line_width,color='#FA7F6F')

axes.tick_params(axis='x',labelsize=font_size)
axes.tick_params(axis='y',labelsize=font_size)
plt.grid(True)
plt.legend(loc='upper right',prop={'size':font_size})
plt.savefig("../images/GA.png",dpi=300)
plt.show()
```
